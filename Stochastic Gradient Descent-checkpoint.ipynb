#### Import numpy, matplotlib.pyplot and make it inline

```python
import numpy as np
from numpy.linalg import norm
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(42)
```

To have a dataset or set of data points, the student must generate a pair of arrays <b>X</b> and <b>y</b> with the values in <b>X</b> equally distributed between <b>0</b> and <b>20</b> and the values in <b>y</b> such that: 
<b>yi = a*xi + b (and a = -2, b = 1)</b>


```python
a = -2
b = 1
X = np.linspace(0,20,50)
y = a * X + b
ones = np.ones((len(X),1))
X = X.reshape((len(X), 1))
X = np.hstack([ones, X])
```

## Stochastic GD

#### Update the implementation of the Batch GD to beStochastic and rpeat the above plots and evaluations
#### You can try different learning rate and see its effect.
#### Compare to Batch and Mini-Batch GD

```python
def stochastic_GD(X, y, alpha, epochs) :
    
    theta = np.random.random(X.shape[1])
    weights = []
    loss = [0]
    count = 0
    for epoch in range(epochs) :     
        for i in range(X.shape[0]) : 
            h_x = theta @ X[i] 
            gradient = 2 * (h_x - y[i]) * X[i]
            theta = theta - alpha * gradient
        h_x = X @ theta    
        loss.append(sum((h_x - y)**2) / (2 * len(y)))       
        weights.append(theta)      
        count += 1
        # stopping criteria
        if norm(gradient, 2) < 0.00001:
            print("Gradient is so small")
            break
        if abs(loss[-2] - loss[-1]) < 0.000001 :
            print("loss is decreasing slowly")
            break
            
    y_pred = X @ theta
    
    return  weights, loss, y_pred, theta, count                 
```

```python
epochs = 10000
alpha = 0.0001

weights, loss, y_pred, theta, count = stochastic_GD(X, y, alpha, epochs)
```

```python
print(count)
print(len(loss) - 1)
print(theta)
```

```python
from sklearn.metrics import r2_score
squared_R = r2_score(y, y_pred)
squared_R
```

```python
iterations = np.arange(0,len(loss) ,1)
plt.plot(iterations, loss)
plt.xlabel("Number Of Iterations")
plt.ylabel("Cost Function")
plt.title("Iterations VS Cost Function");
```

```python
weights = np.array(weights)
plt.plot(weights[ : , 0], loss[1:], label = "theta_0")
plt.plot(weights[ : , 1], loss[1:], label = "theta_1")
plt.title("Weights VS Loss")
plt.xlabel("Theta")
plt.ylabel("Loss")
plt.legend();
```

```python
plt.scatter(X[:, 1], y)
for theta in weights :
    plt.plot(X[:, 1], X @ theta, color = 'r')
    
plt.xlabel("X Points")
plt.ylabel("y Points");
```
