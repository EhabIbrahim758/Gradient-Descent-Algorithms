#### Import numpy, matplotlib.pyplot and make it inline

```python
import numpy as np
from numpy.linalg import norm
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

To have a dataset or set of data points, the student must generate a pair of arrays <b>X</b> and <b>y</b> with the values in <b>X</b> equally distributed between <b>0</b> and <b>20</b> and the values in <b>y</b> such that: 
<b>yi = a*xi + b (and a = -1, b = 2)</b>


```python
a = -1
b = 2
X = np.linspace(0,20,50)
y = a * X + b
ones = np.ones((len(X),1))
X = X.reshape((len(X), 1))
X = np.hstack([ones, X])
```

#### Plot your data points. 

```python
plt.scatter(X[:, 1], y, label = 'Original Data')
```

#### Use your final version of the implemented Batch GD algorithm for single variable to find the best line filt of the given data.
#### Coose the suitable number of iterations, learning rate, and stop criteria.
#### Calculate r2 score. Shouldn't below 0.9
#### Plot the required curves (loss-epochs, loss-theta0, loss-theta1, all fitted lines per epoch (single graph) , best fit line)

```python
def Batch_GD(X, y, alpha, max_iter) :
    theta = np.random.random(X.shape[1])
    
    loss = [0]
    weights = []
    count = 1
    
    for i in range(max_iter) : 
        h_x = X @ theta
        cost = (1 /(2 * len(y))) * sum(np.power(h_x - y, 2))
        error = h_x - y
        gradient = (1 / len(y)) * X.T @ error
        theta = theta - alpha * gradient
        loss.append(cost)
        weights.append(theta)
        count += 1
        if norm(gradient, 2) < 0.0001:
            break
        if abs(loss[-2] - loss[-1]) < 0.0000001 :
            break
        
    y_pred = X @ theta
    
    return  weights, loss, y_pred, theta, count     
```

```python
alpha = 0.008
max_iter = 10000
weights, loss, y_pred, theta, count = Batch_GD(X, y, alpha, max_iter)
```

```python
print(count)
print(len(loss))
print(len(weights))
print(theta)
```

```python
from sklearn.metrics import r2_score
squared_R = r2_score(y, y_pred)
squared_R
```

```python
iterations = np.arange(0,count ,1)
plt.plot(iterations, loss)
plt.xlabel("Number Of Iterations")
plt.ylabel("Cost Function")
plt.title("Iterations VS Cost Function");
```

```python
weights = np.array(weights)
plt.plot(weights[ : , 0], loss[1:], label = "theta_0")
plt.plot(weights[ : , 1], loss[1:], label = "theta_1")
plt.title("Weights VS Loss")
plt.xlabel("Theta")
plt.ylabel("Loss")
plt.legend();
```

```python
plt.scatter(X[:, 1], y)
for theta in weights :
    plt.plot(X[:, 1], X @ theta, color = 'r')
    
plt.xlabel("X Points")
plt.ylabel("y Points");
```

```python
plt.scatter(X[:, 1], y, label = 'Original Data')
plt.plot(X[:, 1], y_pred, color = 'r', label = 'Predicted Data')
plt.legend()
plt.xlabel("X Points")
plt.ylabel("y Points");
```

### Update your Batch GD for one variable  implementation to be Momentum-Based GD and check your results

#### γ takes values between 0 and 1.


```python
def Momentum_Batch_GD(X, y, alpha, gamma ,max_iter) : 
    theta = np.random.random(X.shape[1])
    
    velocity = np.zeros(2)
    
    loss = [0]
    weights = []
    count = 0
    
    for i in range(max_iter) : 
        h_x = X @ theta
        cost = (1 /(2 * len(y))) * sum(np.power(h_x - y, 2))
        error = h_x - y
        gradient = (1 / len(y)) * X.T @ error
        velocity = gamma * velocity + alpha * gradient
        theta = theta - velocity
        loss.append(cost)
        weights.append(theta)
        count += 1
        if norm(gradient, 2) < 0.0001:
            break
        if abs(loss[-2] - loss[-1]) < 0.0000001 :
            break
        
    y_pred = X @ theta
    
    return  weights, loss, y_pred, theta, count     




```

```python
alpha = 0.001
gamma = 0.5
max_iter = 10000

weights, loss, y_pred, theta, count = Momentum_Batch_GD(X, y, alpha, gamma ,max_iter)
```

### Update your Batch GD for one variable  implementation to be NAG and check your results

```python
def Nesterov_GD(X, y, alpha, gamma ,max_iter) : 
    
    theta = np.random.random(X.shape[1])
    velocity = np.zeros(2)    
    
    loss = [0]
    weights = []
    count = 0
    
    for i in range(max_iter) : 
        h_x = X @ theta
        cost = (1 /(2 * len(y))) * sum(np.power(h_x - y, 2))
        error = h_x - y       
        gradient = (1 / len(y)) * X.T @ error
        
        theta_temp = theta - gamma * velocity
        gradient_temp = (1 / len(y)) * ((X @ theta_temp) - y) @ X
        theta = theta_temp - alpha * gradient_temp
        velocity = gamma * velocity + alpha * gradient_temp
                
        loss.append(cost)
        weights.append(theta)
        count += 1
        if norm(gradient, 2) < 0.001:
            break
        if abs(loss[-2] - loss[-1]) < 0.00001 :
            break
        
    y_pred = X @ theta
    
    return  weights, loss, y_pred, theta, count     
```
