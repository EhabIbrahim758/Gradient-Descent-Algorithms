#### Import numpy, matplotlib.pyplot and make it inline

```python
import numpy as np
from numpy.linalg import norm
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(42)
```

To have a dataset or set of data points, the student must generate a pair of arrays <b>X</b> and <b>y</b> with the values in <b>X</b> equally distributed between <b>0</b> and <b>20</b> and the values in <b>y</b> such that: 
<b>yi = a*xi + b (and a = -2, b = 1)</b>


```python
a = -2
b = 1
X = np.linspace(0,20,50)
y = a * X + b

ones = np.ones((len(X),1))
X = X.reshape((len(X), 1))
X = np.hstack([ones, X])
```

```python
def mini_batch_GD(X, y, alpha, epochs, batch_size) :
    #np.random.shuffle(X)
    theta = np.random.random(2)
    weights = []
    loss = [0]
    count = 0
    for epoch in range(epochs) :     
        for i in range(int(X.shape[0] / batch_size)) : 
            X_batch = X[i*batch_size : batch_size * (i+1)]
            y_batch = y[i*batch_size : batch_size * (i+1)]
            h_x = X_batch @ theta
            error = h_x - y_batch
            cost = (1 /(2 * len(y_batch))) * sum(error**2)
            gradient = (1 / len(y_batch)) * X_batch.T @ error
            theta = theta - alpha * gradient
        # storing loss and weights of every batch    
        loss.append(cost)
        weights.append(theta)
        count += 1        
        # stopping criteria
        if norm(gradient, 2) < 0.0001:
            print("gradient is less than 0.001")
            break
        if abs(loss[-2] - loss[-1]) < 0.000001 :
            print("decreasing loss is small")
            break

    y_pred = X @ theta
    
    return  weights, loss, y_pred, theta, count   
```

```python
alpha = 0.000729
epochs = 10000
batch_size = 5
weights, loss, y_pred, theta, count = mini_batch_GD(X, y, alpha, epochs, batch_size)
```

```python
from sklearn.metrics import r2_score
squared_R = r2_score(y, y_pred)
squared_R
```

```python
iterations = np.arange(0,count ,1)
plt.plot(iterations, loss[1:])
plt.xlabel("Number Of Iterations")
plt.ylabel("Cost Function")
plt.title("Iterations VS Cost Function");
```

```python
weights = np.array(weights)
plt.plot(weights[ : , 0], loss[1:], label = "theta_0")
plt.plot(weights[ : , 1], loss[1:], label = "theta_1")
plt.title("Weights VS Loss")
plt.xlabel("Theta")
plt.ylabel("Loss")
plt.legend();
```

```python
plt.scatter(X[:, 1], y)
for theta in weights :
    plt.plot(X[:, 1], X @ theta, color = 'r')
    
plt.xlabel("X Points")
plt.ylabel("y Points");
```
