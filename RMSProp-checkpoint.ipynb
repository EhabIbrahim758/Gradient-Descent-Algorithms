### Imports

```python
import numpy as np
from numpy.linalg import norm
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(42)
```

### getting data

```python
data = pd.read_csv("MultipleLR.csv", header = None)
X = data.iloc[:, 0:3]
y = data.iloc[:, 3]
ones = np.ones((X.shape[0], 1))
X = np.hstack([ones, X])
```

#### the function will take X, y, learning rate, beta, max iterations
#### and return all theta, optimum theta, all losses, number of iterations 


```python
def RMSProp(X, y, alpha, beta, max_iter) :
    # guessing thetas
    theta = np.random.random(X.shape[1])
    
    loss = [0]
    iterations = 0
    weights = []
    v_t = 0

    for i in range(max_iter) :
        h_x = X @ theta
        error = h_x - y
        cost = sum(error**2) / (2 * len(y))
        gradient =  error.T @ X
        v_t = beta * v_t + (1-beta) * gradient**2
        theta = theta - (alpha / np.sqrt(v_t+1e-8)) * gradient
        iterations += 1
        loss.append(cost)
        weights.append(theta)
        
        print(v_t)
        print(gradient)
        print(alpha / np.sqrt(v_t+1e-8))
        
        # stopping condition
        if norm(gradient, 2) < 0.01 :
            break
        if abs(loss[-2] - loss[-1]) < (1e-6) :
            break
        
    y_pred = X @ theta
    return iterations, weights, loss, y_pred 
```

```python
alpha = 0.005
max_iter = 100000
beta = 0.9
iterations, weights, loss, y_pred  = RMSProp(X, y, alpha, beta, max_iter)
```

```python
print(iterations)
print(loss[-1])
print(weights[-1])
```

### the accuracy of the model

```python
from sklearn.metrics import r2_score
squared_R = r2_score(y, y_pred)
squared_R
```

### plotting loss function VS iterations

```python
x_values = np.arange(0, iterations, 1 )
plt.plot(x_values, loss[1:])
#plt.xlim(0, 20000)
#plt.ylim(0, 10)
plt.xlabel("Number Of Iterations")
plt.ylabel("Cost Function")
plt.title("Iterations VS Cost Function");
```
